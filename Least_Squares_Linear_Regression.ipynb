```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Sample Dataset
np.random.seed(0)
x = np.linspace(0, 10, 20)
y = 2.5 * x + 1.0 + np.random.normal(scale=2, size=x.shape)  # true line: y = 2.5x + 1 + noise

# 2. Least Squares Method
# Construct matrices
X = np.vstack([x, np.ones_like(x)]).T  # Design matrix with x and bias term
beta, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)  # Solve for [m, b]

m, b = beta
print(f"Least Squares Solution: y = {m:.2f}x + {b:.2f}")

# 3. Plot Original Data and Best-Fit Line
plt.scatter(x, y, label='Data Points')
plt.plot(x, m * x + b, color='red', label='Least Squares Fit')
plt.title('Least Squares Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
